```python
### How long does it usually take for a company to become a unicorn? Has it always been this way?

import pandas as pd

df = pd.DataFrame({
    'Date': ['22/03/2022', '09/03/2022', '30/03/2022', '07/03/2022', '16/03/2022', '22/03/2022', '17/03/2022', '17/03/2022', '04/03/2022', '09/03/2022', '29/03/2022', '14/03/2022', '10/03/2022'],
    'Founded': [2021, 2012, 2006, 2018, 2015, 2017, 2017, 2019, 2011, 2012, 2018, 2019, 2015]
})

# convert date columns to datetime objects
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Founded'] = pd.to_datetime(df['Founded'], format='%Y')

df['time_to_unicorn'] = df['Date'] - df['Founded']
average_time_to_unicorn = df['time_to_unicorn'].mean().days / 365

print(f"The average time it takes for companies to become unicorns is {average_time_to_unicorn:.1f} years.")
```

```python
## Multivariate Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Load the unicorn dataset
df = pd.read_csv(r'C:\Users\nkech\OneDrive\Documents\Unicorn_Companies.csv')

# Assuming you have a DataFrame 'df' with columns 'Industry' and 'Funding'
df['Valuation'] = df['Valuation'].str.replace('$', '').str.replace('B', '').astype(float)

# Selecting the relevant columns for correlation analysis
selected_columns = ['Funding', 'Valuation']

# Creating a new dataframe with the selected columns
selected_df = df[selected_columns]

# Calculating the correlation matrix
correlation_matrix = selected_df.corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix: Industry vs Valuation')
plt.show()
```

```python
## The above heatmap chart shows there is a positive correlation between industry and valuation, meaning that valuations are made based on the industry type.
```

```python
### companies with highest valuations per industry


import pandas as pd
# Load the unicorn dataset
df = pd.read_csv(r'C:\Users\nkech\OneDrive\Documents\Unicorn_Companies.csv')

# Assuming you have a DataFrame 'df' with columns 'Industry', 'Company', and 'Valuation'

# Convert 'Valuation' column to numeric data type
df['Valuation'] = pd.to_numeric(df['Valuation'].str.replace('$', '').str.replace('B', ''), errors='coerce')

# Group the DataFrame by industry and find the company with the highest valuation within each industry
highest_valuation_companies = df.groupby('Industry').apply(lambda x: x.loc[x['Valuation'].idxmax()])

# Print the companies with the highest valuation per industry
print("Companies with the highest valuation per industry:")
print(highest_valuation_companies[['Industry', 'Company', 'Valuation']])
```

```python
## from the above output, we identify the companies that had the highest valuations in their repsective industries. Other companies in these industries might want to take a look at what these highest valued companies are doing and see if they can identify areas for improvement and optimize business operations. 
- The Artifical Intelligence industry records the company with the highest valuation in the data set while the Travel industry records the least.
```

```python
### the top 50 cities that has unicorn companies with highest valuations


# Load the unicorn dataset
import pandas as pd

unicorn_df = pd.read_csv(r'C:\Users\nkech\OneDrive\Desktop\Unicorn_Companies.csv')

# Convert the Valuation column from ($B) to numeric
unicorn_df["Valuation"] = unicorn_df["Valuation"].str.replace('B','').str.replace('$','').astype(float) * 1000

# Group the dataset by city and country and calculate the mean valuation and year founded for each city-country pair
city_country_df = unicorn_df.groupby(["City", "Country"])[["Valuation", "Founded"]].mean().reset_index()

# Sort the cities by valuation in descending order and select the top fifty cities
top_cities = city_country_df.sort_values("Valuation", ascending=False).head(50)

# Print the top fifty cities with highest valuations, their countries, and year founded
print(top_cities[["City", "Country", "Valuation", "Founded"]])
```
